Identify and remove outliers if any

```python
## using the Z score can measure how many standard devisations a data point is from the mean - scores greater than the threshold can be considered outliers, smaller datasets use 2(95%) 
##and larger datasets use 4 (99%)
```

```python
## make new numerical column variables
num_variables2 = list(df.select_dtypes(include=['int', 'float']))
```

```python
num_variables2
```

```python
## calculate z score for only the numerical columns to prevent error
z_scores= stats.zscore(df[num_variables2])
```

```python
z_scores
```

```python
outliers = (z_scores>3).any(axis=1)  ## can change 3 if need be???
```

```python
outliers.sum()
```

```python
df=df[~outliers]  ## removes rows with outliers. 
# the ~ operator inveertsthe boolean values of the outliers, so the True values become faluse and vice versa
```
